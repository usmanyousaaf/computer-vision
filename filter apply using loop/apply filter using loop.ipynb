```python

if image is None:
    print('Error: Could not read the image')
else:
    # Define the filter kernel
    kernel = np.array([[1, 0, 1],
                       [0, 1, 0],
                       [1, 0, 1]])

    # Create an empty output array
    output = np.zeros_like(image)

    # Loop through each pixel of the image array (excluding the outer border pixels) and apply the filter kernel
    for i in range(1, image.shape[0]-1):
        for j in range(1, image.shape[1]-1):
            # Extract the 3x3 region of the image corresponding to the current pixel for each color channel
            region = image[i-1:i+2, j-1:j+2, :]
            # Apply the filter kernel to the region for each color channel
            filtered_value = np.sum(region * kernel, axis=(0, 1))
            # Store the filtered value in the output array
            output[i-1, j-1, :] = filtered_value

   

```
